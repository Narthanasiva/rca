# **2. Group the dataframe for each device**

*   "deviceId" is used to group

```python
grp_deviceId = df.groupby('deviceId')
```

```python
for id,df in grp_deviceId:
  df_columns = list(df.columns)
  print(df_columns)
  table = df.reset_index().pivot(values='value', index="time", columns='code')
  print(table)
  table.to_csv('/content/drive/MyDrive/Senzmate/RCA/Analyzed_dataframes_for_each_devices/{}.csv'.format(id))
  # break
```
