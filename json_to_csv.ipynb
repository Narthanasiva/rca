```python
import json
import csv
```

```python
def divide_chunks(l, n):
     
    # looping till length l
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
file_name = "/content/drive/MyDrive/Senzmate/RCA/sensor.json"
```

```python
with open(file_name) as json_file:
    jsondata = [json.loads(line) for line in open(file_name,'r')]
num = 1
chunk_data = list(divide_chunks(jsondata, num))
len(jsondata)
data_file = open("/content/drive/MyDrive/Senzmate/RCA/sensor.csv", 'w', newline='')
csv_writer = csv.writer(data_file)
 
count = 0
for chunk in chunk_data:
  for data in chunk:
      data['_id'] = data['_id']['$oid']
      data['time'] = data['time']['$date']
      data['creationDate'] = data['creationDate']['$date']
      data['modifiedDate'] = data['modifiedDate']['$date']
      if count == 0:
          header = data.keys()
          print(header)

          csv_writer.writerow(header)
          count += 1
      csv_writer.writerow(data.values())
 
data_file.close()
```
