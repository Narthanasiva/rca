# **1. Clean Data**

```python
input_file_path = '/content/drive/MyDrive/Senzmate_New/RCA/sensor.csv'

output_file_path = '/content/drive/MyDrive/Senzmate_New/RCA/sensors_cleaned_data.csv'
```

```python
main_df=pd.read_csv(input_file_path,  on_bad_lines='skip', index_col=False) #reading csv file
```

```python
main_df.head(5)
```

```python
main_df = pd.DataFrame(main_df)[['deviceId', 'code', 'time', 'value']]
main_df.head(5)
```

```python
needed_sensor_data = ["B", "IT", "LIA1", "H", "IRO", "T", "ST", "SS"]
```

```python
main_df2 = main_df.loc[main_df['code'].isin(needed_sensor_data)].reset_index(drop=True)
main_df2.head()
```

```python
main_df2.to_csv(output_file_path, index=False)  
```
