# **1. Clean Data**

```python
# input_file_path = '/content/drive/MyDrive/Senzmate/RCA/test_jsonoutput3.csv'
input_file_path = '/content/drive/MyDrive/Senzmate/RCA/jsonoutput3.csv'

# output_file_path = '/content/drive/MyDrive/Senzmate/RCA/results/test_jsonoutput3_cleaned_data.csv'
output_file_path = '/content/drive/MyDrive/Senzmate/RCA/results/jsonoutput3_cleaned_data.csv'
```

```python
main_df=pd.read_csv(input_file_path,  on_bad_lines='skip', index_col=0) #reading csv file
```

```python
main_df.head(5)
```

```python
main_df.drop(['_id', '_class', 'number', 'creationDate', 'modifiedDate'], axis=1, inplace=True)
main_df.head(5)
```

```python
first_column_label = main_df.columns[0]
main_df = main_df.drop(columns=first_column_label)
main_df.head(5)
```

```python
needed_sensor_data = ["B", "IT", "LIA1", "H", "IRO", "T", "ST", "SS"]
```

```python
main_df2 = main_df.loc[main_df['code'].isin(needed_sensor_data)].reset_index(drop=True)
main_df2.head()
```

```python
main_df2.to_csv(output_file_path, index=False)  
```
